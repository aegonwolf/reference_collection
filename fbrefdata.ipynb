```python
def rename_cols(df):
    #drop the league column
    df = df.iloc[:, 1:] 
    df = df.rename(columns={'Unnamed: 1':'season', 'Unnamed: 2':'team', 'Unnamed: 3':'player'})
    cols = []
    for col, i, in zip(df.columns, df.iloc[0]):
        if type(i) == str:
            col = col.split('.')[0].lower() + '_' + i.lower()
        cols.append(col)
    df.columns = cols
    
    return df.drop([0, 1], axis=0).reset_index(drop = True)
```

```python
pposs = rename_cols(pposs)
tposs = rename_cols(tposs)
standardp = rename_cols(standardp)
standardt = rename_cols(standardt)
```

```python
player_sums['xGContrib'] = player_sums['Expected'][['npxG', 'xA']].sum(axis=1)
player_sums['defenseContrib'] = player_sums['Performance'][['Tkl', 'Int']].sum(axis=1)
```

```python
player_sums['games_played'] = player_sums.groupby(['player'])['game_id'].transform('count')
```

```python
maxXGC = player_sums['xGContrib'].max()
maxDFC = player_sums['defenseContrib'].max()
```

```python
player_means = player_sums.reset_index().groupby('player').mean()
```

```python
player_means['normalizedXGC'] = player_means['xGContrib'] / maxXGC
player_means['normalizedDFC'] = player_means['defenseContrib'] / maxDFC
```

```python
player_means['games_played']
```

```python
player_means['mean_contrib'] = (0.5*player_means['normalizedXGC'] + 0.5*player_means['normalizedDFC'])*player_means['games_played']
```
